```python
from PyPDF2 import PdfReader

reader = PdfReader("/Users/ashishvashist/Documents/LLM/Resume_Ashish_Vashist.pdf")
number_of_pages = len(reader.pages)
page = reader.pages[0]
text = page.extract_text()
```

```python
def text_to_json(text):
    prompt = f"""
    Please convert the following resume raw text into a JSON format. Ensure the JSON includes the following keys:
    - name: The name of the candidate.
    - personal_info: Including email, phone, LinkedIn, GitHub.
    - experience: A list of all the experience, including company name, role, and focus areas.
    - projects: A list of projects, including title, description, and technologies used.
    - achievements: A list of achievements.
    - education: A list of educational qualifications, including institution name, degree, duration, and CGPA (if available).

    {text}
    """

    response = openai.ChatCompletion.create(
        model="gpt-4-turbo",  
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=2048
    )

    return response.choices[0].message['content'].strip()
```

```python
openai.api_key = ('OPENAI_API_KEY')
output = text_to_json(text)
print(output)
```
